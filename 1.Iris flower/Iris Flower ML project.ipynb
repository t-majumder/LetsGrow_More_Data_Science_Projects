```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

# Dataset Loading

```python
columns=['sepal length','sepal width','petal length','petal width','flower class name']
df=pd.read_csv('D:\\Data science and ml projects\\1.Iris flower\\iris.data',names=columns)
df.head(150)
```

```python
df.describe()
```

# DATA   VISUALIZATION

```python
#plotting using seaborn
sns.pairplot(df,hue='flower class name')
```

# Training the Model

```python
#Separating features and columns
data=df.values
X=data[:,0:4]
Y=data[:,4]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
#Support vector mechine algorithm
from sklearn.svm import SVC
model_svc=SVC()
model_svc.fit(X_train,Y_train)
```

```python
prediction1=model_svc.predict(X_test)
from sklearn.metrics import accuracy_score
print("The Accuracy of the prediction is: ",accuracy_score(Y_test,prediction1)*100)
```

```python
for i in range(len(prediction1)):
    print(Y_test[i],'    ',prediction1[i])
```

## LINEAR REGRASSION MODEL

```python
#LINEAR REGRASSION
import warnings
warnings.filterwarnings("ignore")
from sklearn.linear_model import LogisticRegression
model_LR=LogisticRegression()
model_LR.fit(X_train,Y_train)
```

```python
prediction2=model_LR.predict(X_test)
from sklearn.metrics import accuracy_score
print("The Accuracy of Linear Regrassion Model is: ", accuracy_score(Y_test,prediction2)*100)
```

## Decision tree Classifier Model

```python
from sklearn.tree import DecisionTreeClassifier
model_DTC=DecisionTreeClassifier()
model_DTC.fit(X_train,Y_train)
```

```python
prediction3=model_DTC.predict(X_test)
from sklearn.metrics import accuracy_score
print("The Accuracy of Decision tree Classifier Model is: ", accuracy_score(Y_test,prediction3)*100)
```

```python
# R  E  P  O  R  T
from sklearn.metrics import classification_report
print(classification_report(Y_test,prediction3))
```
